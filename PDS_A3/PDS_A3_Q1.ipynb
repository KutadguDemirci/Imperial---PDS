# Part C (12 marks) 

Provide implementations for the four classes from part A. All tests from part B should now pass without errors or failures. You may add additional methods if that is helpful. 

```python
# copy your code from Part A here first, then fill in the details while rerunning the tests from Part B repeatedly
# add the interfaces here

#Import the libraries
import pandas as pd
import numpy as np

#Reading the data and checking its columns
trips = pd.read_csv('trips.csv')
print(trips.head())



####################### Part 1 ##############################
class CycleTrip:

#Create the arguments that the class will accept.
    def __init__(self, start_time, end_time, ride_type):

        #Create the private attributes

        # these 2 has to be TimeStamp
        self._start_time = pd.to_datetime(start_time, dayfirst = True)
        self._end_time = pd.to_datetime(end_time, dayfirst = True)

        self._is_electric = (ride_type == 'electric')
        self._ride_type = ride_type

#Create the method for returing bike type
    def is_electric(self):
        return self._is_electric

#Create the method for getting day name
    def get_start_day(self):
        return self._start_time.day_name()

#Create an empty method that will be used later.
    def compute_profit(self):
        return 0.0

#Create the __str__
    def __str__(self):
        # Calculate duration
        duration = self._end_time - self._start_time
        # Call compute_profit method to get the value
        profit = self.compute_profit()
        return f"CycleTrip: {self._ride_type} ride, {duration.seconds / 60} minutes, {profit} profit"



#Trial run
trial = trips.iloc[0]
trip_instance = CycleTrip(trial['started_at'], trial['ended_at'], trial['ride_type'])
print(trip_instance)




####################### Part 2 ##############################
class MemberTrip(CycleTrip):
    def compute_profit(self):

        #calculate trip duration
        trip_minutes = (self._end_time - self._start_time).seconds / 60

        #no profit for trips under 45mn
        if trip_minutes <= 45:
            return 0.0
        
        #calculate overage minutes and profit
        overage = trip_minutes - 45

        if self._is_electric:
            profit = overage * 0.15
        else:
            profit = overage * 0.1

        return profit

####################### Part 3 ##############################
class CasualTrip(CycleTrip):
    def compute_profit(self):

        #calculate trip duration
        trip_minutes = (self._end_time - self._start_time).seconds / 60

        if self._is_electric:
            profit = trip_minutes * 0.4
        else:
            profit = trip_minutes * 0.25

        return profit

####################### Part 4 ##############################

class CycleTripCollection:
    def __init__(self):
        self._trips = []
        self._daily_profits = {}
    
    def add(self, trip):
        #add the trip to the list
        self._trips.append(trip)
        
        # Update daily profits
        day = trip.get_start_day()
        profit = trip.compute_profit()
        
        if day in self._daily_profits:
            self._daily_profits[day] += profit
        else:
            self._daily_profits[day] = profit
    
    def get_trips(self):
        return self._trips
    
    def get_daily_profits(self):
        return self._daily_profits
    
    def __str__(self):
        total_trips = len(self._trips)
        electric_trips = sum(1 for trip in self._trips if trip.is_electric())
        
        return (f"CycleTripCollection\n"
                f" -{total_trips} trips total, {electric_trips} electric trips,\n"
                f" -Daily profits: {self._daily_profits}")
```
